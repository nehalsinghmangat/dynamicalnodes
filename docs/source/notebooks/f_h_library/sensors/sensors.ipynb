# Sensors

Sensors are **stateless** `DynamicalSystem` blocks â€” no `f`; only `h`.
`h` extracts the relevant channels from the state vector, adds Gaussian noise,
and returns an array that maps directly to a ROS message via a `py2ros_*` converter.

## GPS

GNSS receiver outputting latitude, longitude, and altitude.
Stateless â€” only `gps_h` is used. See the [gps](../notebooks/gps.ipynb) notebook for derivation, API reference, and examples.

```python
import numpy as np
from jaxtyping import Float, jaxtyped
from beartype import beartype


@jaxtyped(typechecker=beartype)
def gps_h(*args, **kwargs) -> Float[np.ndarray, "3"]:
    """GPS sensor model.

    Returns:
        ndarray(3,) â€” [latitude (degrees), longitude (degrees), altitude (metres)].
        Layout matches py2ros_nav_sat_fix exactly, so the output can be passed
        directly to that converter to produce a sensor_msgs/NavSatFix message.
    """
    raise NotImplementedError("You need to implement me!")
```

### Ex: `DynamicalSystem` $\to$ `ROSNode`

```python
from dynamicalnodes import DynamicalSystem, ROSNode
from dynamicalnodes.ros2py_py2ros import py2ros_nav_sat_fix
from sensor_msgs.msg import NavSatFix

gps = DynamicalSystem(h=gps_h)
gps_node = ROSNode(
    dynamical_system=gps,
    publishes_to=[
        {
            "topic": "/gps",
            "msg_type": NavSatFix,
            "py2ros": py2ros_nav_sat_fix,
        },  # (3,) â†’ NavSatFix
    ],
    timer_hz=1,
)
```
